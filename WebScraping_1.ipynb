```python
csv_file = open('testHotel.csv', 'w')
csv_writter = csv.writer(csv_file)
csv_writter.writerow(['hotelName', 'hotelAddress', 'hotelRating', 'hotelReview','hotelDescription', 'hotelPrice','hotelDistance(Km)', 'hotelPicsUrls', 'hotelSource'])
# csv_file.close()
```

```python
number = 0
for item in soup.select('.bui-carousel__item'): 
    try:
       number  = number + 1
       print(number)

       hotelName = (item.select('.bui-card__title')[0].get_text().strip())
       hotelAddress = (item.select('.bui-card__subtitle')[0].get_text().strip()) 
       hotelRating = '5 Star'
       hotelReview = (item.select('.review-score-badge')[0].get_text().strip()) 
       hotelDescription = (item.select('.bui-card__text')[0].get_text().strip())
       hotelPrice = (item.select('.hotel-card__price')[0].get_text().strip().replace('From', '')) 
       hotelDistance = '1'
       hotelPicsUrls = 'https://www.booking.com'+ (item.select('.bui-card__header_full_link_wrap')[0]['href'])
       hotelSource = 'booking.com'

       csv_writter.writerow ([hotelName, hotelAddress, hotelRating, hotelReview, hotelDescription, hotelPrice.replace(' per night', ''), hotelDistance,hotelPicsUrls, hotelSource ])

      # Print The Data
      # Print The Data
       print (hotelName)                 # 1 Hotel name
       print (hotelAddress)              # 2 Address
       print (hotelRating)               # 3 Hotel Rating
       print (hotelReview)               # 4 Hotel Review
       print (hotelDescription)          # 5 Hotel Description - Short 
       print (hotelPrice)                # 6 Hotel Price Per Night 
       print (hotelDistance)             # 7 Hotel Distance
       print (hotelPicsUrls)             # 8 Hotel Page Full Link Need to append https://www.booking.com/<Link
       print (hotelSource)               # 9 Hotel Source

       print('---------------------------')
    except Exception as e: print('NA') 

# csv_file.close()
```
